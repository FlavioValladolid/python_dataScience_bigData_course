# Etiquetas y leyendas en Matplotlib

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('base_datos_2008.csv', nrows=100000)
data = np.unique(df.DayOfWeek, return_counts = True) # Crea una tabla de resumen, numero de vuelos que han salido cada dia de la semana
labs = ["Lun","Mar","Mie","Jue","Vie","Sab","Dom"] # Definir etiqueta de los datos
data
```

### Realizar un grafico de sectores

```python
plt.pie(x = data[1],
        labels = data[0],
        radius = 1.5,
        colors = ["Red","Green","Orange","Blue","Gray","Pink","Black"],
        startangle = 90)
plt.show()
```

```python
# Modificar etiquetas
plt.pie(x = data[1],
        labels = labs, 
        radius = 1.5,
        colors = ["Red","Green","Orange","Blue","Gray","Pink","Black"],
        startangle = 90)
plt.show()
```

```python
# Usar una paleta de colores predefinida de Seaborn

plt.pie(x = data[1],
        labels = labs, 
        radius = 1.5,
        colors = sns.color_palette("hls",7), # Se indica la cantidad de colores que se necesitan
        startangle = 90)
plt.show()
```

```python
# Separar una rebanada
plt.pie(x = data[1],
        labels = labs, 
        radius = 1.5,
        colors = sns.color_palette("hls",7), 
        explode = (0,0,0,0,0,0,0.1), # Separa un poco el valor 7
        startangle = 90)
plt.show()
```

```python
# Mostrar porcentaje de cada uno de los grupos

plt.pie(x = data[1],
        labels = labs, 
        radius = 1.5,
        colors = sns.color_palette("hls",7), 
        explode = (0,0,0,0,0,0,0.1),
        startangle = 90,
        autopct ="%1.1f%%") # Agregar porcentajes
plt.show()
```

```python
# Ajusta la distancia de las etiquetas
plt.pie(x = data[1],
        labels = labs, 
        radius = 1.5,
        colors = sns.color_palette("hls",7), 
        explode = (0,0,0,0,0,0,0.1),
        startangle = 90,
        autopct ="%1.1f%%",
        labeldistance = 1.5) # Separacion de etiquetas
plt.show()
```

```python
# Aadir una leyenda la grfico
plt.pie(x = data[1],
        #labels = labs, 
        radius = 0.7, # Modificar para evitar translapamiento
        colors = sns.color_palette("hls",7), 
        explode = (0,0,0,0,0,0,0.1),
        startangle = 90,
        autopct ="%1.1f%%")

plt.legend(loc = "upper left", labels = labs)# Aadir leyenda, con "best" selecciona el mejor lugar para ponerla
# Al agregar un parametro sin sentido en loc, python te recomienda parametros
plt.show()
```

### Crear diagrama de barras con el paquete de Seaborn

```python
plt = sns.barplot(x = labs, y = data[1])
plt.set(xlabel = "Da de la semana", ylabel = "Nmero de Vuelos") # No hace falta plot.show()
```
