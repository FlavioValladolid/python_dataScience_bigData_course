# Introducción a Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('base_datos_2008.csv',nrows=100000)
```

### Crear una tabla resumen de los vuelos cancelados

```python
data = np.unique(df.Cancelled,return_counts = True) 
data # Tabla de resumen de todos los casos que tenemos 0 y 1, y cuántas veces aparecen en nuestra base de datos, array[cantidad, cantidad]
```

### Crear un diagrama de sectores

```python
plt.pie(x = data[1],
        labels = data[0], # Asignar etiquetas
        colors = ["Red","Green"], #Cambiar colores
        shadow = True, # Agregar sombra
        startangle = 90, # Cambiar donde empieza los angulos
        radius = 2 ) # Cambiar el radio 
plt.show() # Mostrar
```
