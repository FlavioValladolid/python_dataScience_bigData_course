# Modificar elementos del gráfico en Matplotlib

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

np.random.seed(0) # Semilla por si se quieren replicar los resultados
df = pd.read_csv("base_datos_2008.csv",nrows=1000000)
df = df.sample(frac = 1).head(100) # Remezclar los datos y mostrar los primeros 100
```

### Crear gráfico de burbujas

```python
plt.scatter(x = df.DayofMonth, y = df.ArrDelay, s = df.Distance) # Dependera de tres parametros que seran las columnas del DataFrame, (dia del mes, retraso, distancia recorrida)

```

### Añadir nuevos parámetros
**Alpha:** Es un parámetro que hace que las burbujas sean traslucidas, opacas (=1), transparentes (=0)

Filtro para saber que dias son fines de semana ==> c = df.DayOfWeek.isin([6,7]) 6 y 7 son sabado y domingo

```python
plt.scatter(x = df.DayofMonth, y = df.ArrDelay, s = df.Distance, alpha = 0.3) 
```

```python
plt.scatter(x = df.DayofMonth, y = df.ArrDelay, s = df.Distance, alpha = 0.3, c = df.DayOfWeek.isin([6,7])) 

# Se muestrna dos colores por proporcion, morado son los dias entre semana y los amarillos son aquellos correspondientes a los fines de semana
```

```python
# Agregar titulo
plt.scatter(x = df.DayofMonth, y = df.ArrDelay, s = df.Distance, alpha = 0.3, c = df.DayOfWeek.isin([6,7])) 
plt.title("Retrasos en EEUU")
# Agregar etiquetas en los ejes
plt.xlabel("Día del mes")
plt.ylabel("Retraso al Llegar")
# Añadir limites en el rango de y
plt.ylim([0,150])
# Cambiar el número de marcas que se tiene en el eje x
plt.xticks([0,15,30])
# Añadir texto con ciertas coordenadas
plt.text(x = 28, y = 120, s = 'Mi vuelo')

plt.show() # Por convencion se pone pero no es necesario
```
