# Histogramas y box plots en Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv ('base_datos_2008.csv')
```

```python
df.dropna(inplace = True, subset = ["ArrDelay","DepDelay","Distance"]) # inplace igual a true para no tener que igualarlo, un subset de las variables que nos va a interesar estudiar
```

### Crear gr√°fica

```python
sns.displot(df['Distance']) # Superposicion del histograma mas clasico con una aproximacion de la densidad
```

### Grafico sin la aproximacion de la densidad

```python
sns.displot(df['Distance'], kde = False)
```

### Modificar el numero de bins que tiene el histograma

```python
sns.displot(df['Distance'], kde = False, bins = 100) # Se puede ver de forma mas sencilla la distribucion de los daotsd
```

### Comparar dos distribuciones

```python
sns.kdeplot(df['ArrDelay'])
# A√±adir otro
sns.kdeplot(df['DepDelay'])
# Acotar los ejes
plt.xlim([-300,300]) # Se recomienda verlo con la aproximacion de la densidad que con el histograma en si
```

### Realizar un diagrama de caja clasico

```python
# Crear un nuevo DataFrame con los vuelos que hayan salido de ATL, HOU e IND, seleccionando una sola fraccion de ellos
df2 = df[df['Origin'].isin(['ATL','HOU','IND'])].sample(frac = 1).head(500)
```

```python
sns.boxplot(x='DepDelay',y='Origin',data = df2) # Se observa que hay muchos outliers en estas tres destinaciones
# Se puede filtrar el eje
plt.xlim([-20,150])
# La linea dentro de la caja delimita la mediana, que es el estadistico que separa el 50% inferior de los casos de los 50% superior de los casos
# La linea de lado izquierdo de la caja, antes de que se coloree de azul es el primer cuartil que separa el 25% inferior de los casos del 75% superior de los casos
# La linea a la derecha de la caja, es el tercer cuartil que separa el 75% de los casos del 25% de los casos
# Los rombos son los outliers que detecta boxplot
# El criterio de desicion de estos puntos es todos aquellos que superen 1.5 veces el rango intercuartilico
# El rango intercuartilico es la distancia entre el tercer cuartil y el primer cuartil 
```

```python
Los tipos de graficas mostradas sirven para mostrar los concentrados de nuestros datos y sacar conclusioens exploratorias de como estan distribuido
```
